```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
breast_cancer_data = pd.read_csv('/Users/tessa/Downloads/breast-cancer.data')
breast_cancer_data.columns = ['Class','Age','Menopause','Tumor-si','Inv-Nodes','Node-Caps','Deg-Malig','Breast','Breast-Quad','Irradiat']
print(breast_cancer_data)
```

```python
melted_copy = breast_cancer_data.copy()

melted_data = melted_copy.melt(id_vars='Breast',value_vars='Age',value_name='Age Range') 
#Melted the data to be sorted by Breast, with the variable being age and the value giving the age range.  

melted_data
```

```python
copy_bc = breast_cancer_data.copy() #Create copy of dataframe

copy_bc['seq']=copy_bc.groupby(['Breast-Quad','Tumor-si']).cumcount()   
#Create a sequence to avoid duplicate values while using pivot

pivot_data = copy_bc.pivot(index=['seq','Breast-Quad'],columns='Tumor-si', values=['Deg-Malig']) 
#Pivot table to have Breast-quadrants displayed in the rows, with the columns for tumor-size 
#and the values the corresponding degenterative malignancy.

pivot_data
```

```python
iter_copy_bc = breast_cancer_data.copy() #Create copy of dataframe

for i in range(len(iter_copy_bc)):                    
    #Created loop to iterate through breast cancer data to find the cases that had done irradiation
    if iter_copy_bc.loc[i,'Irradiat'] == 'yes':
        print([iter_copy_bc.loc[i]])
```

```python
aggregate_copy = breast_cancer_data.copy()

aggregate_data = aggregate_copy.groupby(['Tumor-si','Inv-Nodes','Class']).agg({'Class':'count'})
#Grouped tumor-size, nodes and class then aggregated the count of the class types in relation to tumor-size and nodes

aggregate_data
```

```python
copy_group = breast_cancer_data.copy()

groupby_data = copy_group.groupby('Menopause')['Menopause'].count()
#Grouped by menopause stage and counted the occurrences of each stage.

groupby_data
```
